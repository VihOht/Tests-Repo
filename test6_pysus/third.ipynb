```python
from pysus import CNES

cnes = CNES()
cnes.load()
cnes.content
```

```python

files1 = cnes.get_files("LT", uf="SP", month=1, year=2024)
files1
```

```python
paquet = cnes.download(files1)
df1 = paquet.to_dataframe()
df1.columns
```

```python
from pysus import SIH

sih = SIH()
sih.load()
sih.content
```

```python
files2 = sih.get_files("RD", uf="SP", month=1, year=2024)
files2
```

```python
paquet = sih.download(files2)
df2 = paquet.to_dataframe()
df2.columns
```

```python
print(df1.columns)
print(df2.columns)

df2["UF_ZI"]
```

```python
Q_leitos = int(df1["QT_EXIST"].astype(int).sum())
print("Quantidade de leitos: ", Q_leitos)
dias_totais_disponiveis = Q_leitos * 31
print("Dias totais disponiveis no mes: ", dias_totais_disponiveis)

internacoes = df2["DIAS_PERM"].astype(int).sum()
print("Total de dias de internacao no mes: ", internacoes)

print("Taxa de ocupacao: ", internacoes / dias_totais_disponiveis)
```
