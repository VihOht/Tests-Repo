```python
# pip install pysus
from pysus.ftp.datacatalog import CNES
import pandas as pd

cnes = CNES()

# Baixa Leitos (LT) e Estabelecimentos (ST)
df_lt = cnes.get_table("LT")    # Leitos
df_st = cnes.get_table("ST")    # Estabelecimentos (contém NOME)

# Checar nomes de coluna comuns
print("LT cols:", df_lt.columns.tolist()[:30])
print("ST cols:", df_st.columns.tolist()[:30])
```
